```python
# Polynomial Regression
# Importing the Essential Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Housing.csv')
X = dataset.iloc[:, [1]].values
Y = dataset.iloc[:, 0].values
```

# Splitting Data into Train and Test

```python
def test_train_split(data,target,split_perc):
  perm=np.random.permutation(len(data)) #to get random permutation of 0 to len-1
  x_shuff=data[perm]    #obtain shuffled data
  y_shuff=target[perm]    #obtain shuffled targets
  #get the index till which data is in train set
  n=round((((100-split_perc)*0.01))*len(data)) #round is used to round off to the nearest integer
  x_train=x_shuff[:n] #take starting n elements of shuffled data
  y_train=y_shuff[:n]
  x_test=x_shuff[n:] #take remaining elements of shuffled data
  y_test=y_shuff[n:]
  return x_train, y_train, x_test, y_test

x_train,y_train,x_test,y_test=test_train_split(X,Y,20)
```

# Using Inbuilt Sklearn Function


```python
# Fitting Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly_reg_4 = PolynomialFeatures(degree = 4)
X_poly = poly_reg_4.fit_transform(x_train)
poly_reg_4.fit(X_poly, y_train)
lin_reg_4 = LinearRegression()
lin_reg_4.fit(X_poly, y_train)

poly_reg_6 = PolynomialFeatures(degree = 6)
X_poly = poly_reg_6.fit_transform(x_train)
poly_reg_6.fit(X_poly, y_train)
lin_reg_6 = LinearRegression()
lin_reg_6.fit(X_poly, y_train)

poly_reg_8 = PolynomialFeatures(degree = 8)
X_poly = poly_reg_8.fit_transform(x_train)
poly_reg_8.fit(X_poly, y_train)
lin_reg_8 = LinearRegression()
lin_reg_8.fit(X_poly, y_train)

poly_reg_10 = PolynomialFeatures(degree = 10)
X_poly = poly_reg_10.fit_transform(x_train)
poly_reg_10.fit(X_poly, y_train)
lin_reg_10 = LinearRegression()
lin_reg_10.fit(X_poly, y_train)

poly_reg_12 = PolynomialFeatures(degree = 12)
X_poly = poly_reg_12.fit_transform(x_train)
poly_reg_12.fit(X_poly, y_train)
lin_reg_12 = LinearRegression()
lin_reg_12.fit(X_poly, y_train)

poly_reg_14 = PolynomialFeatures(degree = 14)
X_poly = poly_reg_14.fit_transform(x_train)
poly_reg_14.fit(X_poly, y_train)
lin_reg_14 = LinearRegression()
lin_reg_14.fit(X_poly, y_train)
```

```python
# Visualising the Polynomial Regression results
X_grid = np.arange(min(x_train), max(x_train), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(x_train, y_train, color = 'red')
plt.plot(X_grid, lin_reg_4.predict(poly_reg_4.fit_transform(X_grid)), color = 'blue',label="4 degree")
plt.plot(X_grid, lin_reg_6.predict(poly_reg_6.fit_transform(X_grid)), color = 'red',label="6 degree")
plt.plot(X_grid, lin_reg_8.predict(poly_reg_8.fit_transform(X_grid)), color = 'green',label="8 degree")
plt.plot(X_grid, lin_reg_10.predict(poly_reg_10.fit_transform(X_grid)), color = 'cyan',label="10 degree")
plt.plot(X_grid, lin_reg_12.predict(poly_reg_12.fit_transform(X_grid)), color = 'yellow',label="12 degree")
plt.plot(X_grid, lin_reg_14.predict(poly_reg_14.fit_transform(X_grid)), color = 'brown',label="14 degree")
plt.title('Polynomial Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

# Error Metrics

```python
print(f"R2 Score of 4 degree Polynomial is: {lin_reg_4.score(poly_reg_4.fit_transform(x_test), y_test)}")
print(f"R2 Score of 4 degree Polynomial is: {lin_reg_6.score(poly_reg_6.fit_transform(x_test), y_test)}")
print(f"R2 Score of 4 degree Polynomial is: {lin_reg_8.score(poly_reg_8.fit_transform(x_test), y_test)}")
print(f"R2 Score of 4 degree Polynomial is: {lin_reg_10.score(poly_reg_10.fit_transform(x_test), y_test)}")
print(f"R2 Score of 4 degree Polynomial is: {lin_reg_12.score(poly_reg_12.fit_transform(x_test), y_test)}")
print(f"R2 Score of 4 degree Polynomial is: {lin_reg_14.score(poly_reg_14.fit_transform(x_test), y_test)}")
```
