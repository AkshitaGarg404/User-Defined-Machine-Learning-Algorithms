```python
import numpy as np
import matplotlib.pyplot as plt

#create a function to return f(x)=univariate gaussian distribution function
def univariate_gaussian(x, mean, variance):
  exp_term = -0.5 * ((x - mean) ** 2) / variance
  coeff= 1 / np.sqrt(2 * np.pi * variance)
  return coeff * np.exp(exp_term)

#function for plot using matplotlib
def plot_univariate_gaussian(mean, variance):
  x = np.linspace(mean - 3 * np.sqrt(variance), mean + 3 * np.sqrt(variance), 1000)
  prob_fn = univariate_gaussian(x, mean, variance)
  plt.plot(x, prob_fn , label=f"Mean = {mean}, Variance = {variance}")
  plt.xlabel('Random variable: x')
  plt.ylabel('Probability Density')
  plt.title('Univariate Gaussian Distribution')
  plt.legend()
  plt.grid(True)
  plt.show()


mean = [1,2]
variance = [1,2]
#showing variation in graph due to changes in mean and variance
for i in mean:
  for j in variance:
    plot_univariate_gaussian(i, j)
```
