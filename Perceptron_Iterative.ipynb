```python
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
import random
X,Y =make_blobs(n_samples=200,centers=2,cluster_std=1.3)
Y=Y.reshape(200,1)
# print(X)
# print(Y)
plt.scatter(X[:,0],X[:,1],c=Y,cmap='winter') #X[i][j].... X[i,j]
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Raw Data')
plt.show()
```

```python
w_pred=np.random.randint(low=1,high=5,size=(2,1))
b_pred=random.randint(1,5)
#print(bias)
x_pred=np.linspace(-10,10,10)
s_pred=-w_pred[0,0]/w_pred[1,0]
y_inter_pred=-b_pred/w_pred[1,0]
y_pred=s_pred*x_pred+y_inter_pred

x_raw=x_pred
y_raw=y_pred

plt.scatter(X[:,0],X[:,1],c=Y,cmap='winter',edgecolors='k')
plt.plot(x_pred,y_pred,linestyle='dotted',color='red',label='Raw line')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Raw Data')
plt.legend()
plt.show()
print(Y.shape)
```

```python
def perceptron(X,Y,w,b):
  samples,features=X.shape
  y_pred=np.dot(X,w)+b
  count=0;
  for i in range(samples):
    if(y_pred[i]>0): #if positive... above line class1
      y_pred[i]=1
    else:
      y_pred[i]=0
  flag=1
  while flag:
    flag=0
    count+=1
    for i in range(samples):
      if(y_pred[i]!=Y[i]):
        flag=1
        if(Y[i]==0):
          w[0]=w[0]-X[i][0]
          w[1]=w[1]-X[i][1]
          b=b-1
        else:
          w[0]=w[0]+X[i][0]
          w[1]=w[1]+X[i][1]
          b=b+1
        y_pred=np.dot(X,w)+b
      for i in range(samples):
        if(y_pred[i]>0):
          y_pred[i]=1
        else:
          y_pred[i]=0
  return count,w,b


count,w_ans,b_ans=perceptron(X,Y,w_pred,b_pred)
x_pred=np.linspace(-10,10,10)
s_pred=-w_ans[0,0]/w_ans[1,0]
y_inter_pred=-b_ans/w_ans[1,0]
y_pred=s_pred*x_pred+y_inter_pred

plt.scatter(X[:,0],X[:,1],c=Y,cmap='winter',edgecolors='k')
plt.plot(x_pred,y_pred,'-',color='red',label='Predicted')
plt.plot(x_raw,y_raw,linestyle='dotted',color='hotpink',label='Raw line')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Raw Data')
plt.legend()
plt.show()
print(count)
```
